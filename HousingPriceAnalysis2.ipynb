- A skewness of 0 means the data is perfectly symmetrical (like a bell curve).
- A positive skewness means the tail of the distribution is longer on the right side. This often happens when there are a few very high values.
- A negative skewness means the tail of the distribution is longer on the left side. This often happens when there are a few very low values.

```python
train_data.select_dtypes(include='number').skew()
```

Transformation is a technique applied mathematical function to the data to change its distribution and make it more symmetrical, closer to a normal distribution. The logarithmic transformation (taking the logarithm of the values) is a very common and effective way to reduce right skewness. It compresses the larger values more than the smaller values, pulling in that long tail on the right.

For example, if the values are like 1, 10, 100, 1000, the logarithms might be 0, 1, 2, 3 (using log base 10). The large differences between 10, 100, and 1000 become smaller, more manageable differences in the log scale.

```python

if 'SalePrice' in train_data.columns:
    train_data['SalePrice_Log'] = np.log1p(train_data['SalePrice'])


    print("Skewness of original SalePrice:", train_data['SalePrice'].skew())
    print("Skewness of SalePrice_Log:", train_data['SalePrice_Log'].skew())
else:
    print("SalePrice column not found in the DataFrame.")
```

Here I am dropping columns with large number of
missing columns. But first I check the percentage of missing values using .isnull().ismean


```python
missing_pct = train_data.isnull().mean()
print(missing_pct[missing_pct > 0])
```

I then remove columns with high missingness threshold of about over 0.4


```python
threshold = 0.4
missing_pct = train_data.isnull().mean()
drop_missing_val_cols = missing_pct[missing_pct > threshold].index.tolist()
train_data.drop(columns=drop_missing_val_cols, inplace=True)
```

```python
train_data.info()
```

Now I fill in the
missing values in the remaining columns

```python
object_columns = train_data.select_dtypes(include=['object']).columns.tolist()
print(object_columns)
```

Fill in the missing values
for numerical columns I will use median
and for categorical columns i will use mode

Use either inplace= True or assign train_data[col].fillna(median_val) to dropped_cols_data for it to work

```python

numeric_cols = train_data.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = train_data.select_dtypes(include=['object']).columns

for col in numeric_cols:

    if col in train_data.columns:
        median_val = train_data[col].median()
        train_data[col].fillna(median_val, inplace=True)

for col in categorical_cols:

    if col in train_data.columns:
        mode_val = train_data[col].mode()[0]
        train_data[col].fillna(mode_val, inplace=True)
```

```python
target_cols = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']

for col in target_cols:
    if col in train_data.columns:
        median_val = train_data[col].median()
        train_data[col].fillna(median_val, inplace=True)
```

drop columns with high correlation matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Select numeric columns
numeric_dropped_cols_data =train_data.select_dtypes(include=['int64', 'float64'])

# 2. Calculate correlation matrix
corr_matrix = numeric_dropped_cols_data.corr()

# 3. Create upper triangle matrix to avoid duplicate checks
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# 4. Find columns with correlation > 0.9
high_corr_pairs = [(col, idx, corr_matrix.loc[col, idx])
                   for col in upper_triangle.columns
                   for idx in upper_triangle.index
                   if (upper_triangle.loc[idx, col] > 0.8)]

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


subset_features = [
    'SalePrice', 'OverallQual', 'OverallCond',
    'YearBuilt', 'YearRemodAdd', 'YrSold',
    'GrLivArea', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF',
    'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF',
    'FullBath', 'BsmtHalfBath',
    'TotRmsAbvGrd', 'BedroomAbvGr', 'KitchenAbvGr',

]
# Filter to include only columns that exist in the current train_data
subset_features = [f for f in subset_features if f in train_data.columns]
# Makes sure SalePrice is included for the correlation matrix
if 'SalePrice' not in subset_features and 'SalePrice' in train_data.columns:
    subset_features.append('SalePrice')


# Selects only the numeric columns from train_data for correlation
numeric_train_data = train_data[subset_features].select_dtypes(include=['int64', 'float64'])


# Calculates correlation matrix
corr_subset = numeric_train_data.corr()

# Creates upper triangle matrix to avoid duplicate checks (optional for heatmap visualization)
upper_triangle = corr_subset.where(np.triu(np.ones(corr_subset.shape), k=1).astype(bool))

# Finds columns with correlation > 0.9 (optional, can be done visually from heatmap)
high_corr_pairs = [(col, idx, corr_subset.loc[col, idx])
                   for col in upper_triangle.columns
                   for idx in upper_triangle.index
                   if (abs(upper_triangle.loc[idx, col]) > 0.8)]


plt.figure(figsize=(12,10))
sns.heatmap(corr_subset, annot=True, fmt='.2f', cmap='PiYG', cbar=True)
plt.title('Correlation Matrix Heatmap (Subset of Numeric Features)')
plt.show()
```

3. Line Chart: Average SalePrice over YrSold (Year Sold)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

if 'YrSold' in train_data.columns and 'SalePrice' in train_data.columns:
    avg_price_by_year = train_data.groupby('YrSold')['SalePrice'].mean().reset_index()
    plt.figure(figsize=(8,5))
    sns.lineplot(x='YrSold', y='SalePrice', data=avg_price_by_year, marker='o')
    plt.title('Average Sale Price by Year Sold')
    plt.xlabel('Year Sold')
    plt.ylabel('Average Sale Price')
    plt.tight_layout()
    plt.show()
else:
     print("YrSold or SalePrice column not found in the DataFrame.")
```

4. Pie Chart: Proportion by Neighborhood (top 5 neighborhoods, others combined)

```python
import matplotlib.pyplot as plt
import pandas as pd

if 'Neighborhood' in train_data.columns:

    neighborhood_counts = train_data['Neighborhood'].value_counts()


    top_neighborhoods_encoded = neighborhood_counts.nlargest(5)
    others_count_encoded = neighborhood_counts.sum() - top_neighborhoods_encoded.sum()
    pie_data_encoded =  pd.concat([top_neighborhoods_encoded, pd.Series({'Others': others_count_encoded})])

    plt.figure(figsize=(7,7))

    plt.pie(pie_data_encoded, labels=pie_data_encoded.index, autopct='%1.1f%%', startangle=140)
    plt.title('Proportion of Houses by Neighborhood (Encoded Labels)')
    plt.axis('equal')
    plt.show()
else:
    print("Neighborhood column not found in the DataFrame.")
```
